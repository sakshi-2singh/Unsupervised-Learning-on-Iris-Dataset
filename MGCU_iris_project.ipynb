### All together

````python
import pandas as pd
import numpy as np
from  sklearn.linear_model import LogisticRegression
url ='https://raw.githubusercontent.com/sarwansingh/Python/master/ClassExamples/data/iris.csv'

namelist = ['sepal_length',	'sepal_width',	'petal_length',	'petal_width',	'species']
iris = pd.read_csv(url , header=None,names = namelist)

iris1 = np.array(iris)
X = iris1[:,0:4]
Y = iris1[:,4]

model = LogisticRegression()
model.fit(X,Y)
print ( model.predict([[5.1,3.5,1.4,0.2]])[0] )
print ( model.predict([[6.5,2.8,4.6,1.5]] )[0] )
print ( model.predict([[6.5,3.0,5.2,2.0]] )[0] )
````
